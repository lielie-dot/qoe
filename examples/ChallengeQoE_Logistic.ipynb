```python
# Importing the necessary libraries

# Importing data with pandas
import pandas as pd

# Importing numpy to handle numpy data
import numpy as np

# Importing the convenient module train_test_split for data splitting
from sklearn.model_selection import train_test_split

# Importing the library for undersampling
#from imblearn.under_sampling import RandomUnderSampler

# Importing the module for data standardization
from sklearn.preprocessing import StandardScaler

# Importing the module for logistic regression
from sklearn.linear_model import LogisticRegression

# Importing the module for drawing histograms
import matplotlib.pyplot as plt
import seaborn as sns

# Importing the module for creating accuracy scores
from sklearn.metrics import accuracy_score

# Importing the module for creating confusion matrices
from sklearn.metrics import confusion_matrix

# Importing the module for creating ROC curves
from sklearn.metrics import roc_curve
from sklearn.metrics import auc
```

```python
# importation des données
Base_logistic = pd.read_csv("/Users/diao028606/Documents/2022/UC/Challenge QoE/Base/V2/table_matrice_challengeQOEV2.csv", header = 0, sep = ";")
Base_logistic.head(20)
```

```python
# description des données
Base_logistic.describe()
```

```python
# Separating the independent variables (features) and the dependent variable (target)

columnList = Base_logistic.columns.to_list()
columnList.remove("msisdn")
columnList.remove("Cluster")
columnList.remove("test")
columnList.remove("nb_call_est")
columnList.remove("Cible")
X = pd.DataFrame(Base_logistic.loc[:, columnList])
y = pd.DataFrame(Base_logistic.loc[:, ["Cible"]])

# Splitting the data into training and testing sets (training:testing = 7:3)
(X_train, X_test, y_train, y_test) = train_test_split(X, y, test_size=0.3)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
y
```

```python
# Performing logistic regression analysis and building the classification model

# Creating an instance of the LogisticRegression model with a specified parameter
clf = LogisticRegression(C=1000)

# Fitting the model to the training data
clf.fit(X_train, y_train)
```

```python
# Obtaining the coefficients of the classification model

i = 0
for val in columnList:
    if i > 0:
        preStr = " + "
    else:
        preStr = ""
    print(preStr + str(round(clf.coef_[0][i], 4)) + " * " + val)
    i = i + 1
print(" + " + str(clf.intercept_) + " (intercept)")
```

```python
# Performing accuracy validation on the training data

print(accuracy_score(y_train, clf.predict(X_train)))

# Creating a confusion matrix for the training data
conf_mtrx_df = pd.DataFrame(confusion_matrix(y_train, clf.predict(X_train).reshape(-1, 1)))
conf_mtrx_df.rename(columns={0: 'Predicted (0)', 1: 'Predicted (1)'}, index={0: 'Target (0)', 1: 'Target (1)'})
```

```python
# Performing accuracy validation on the testing data

print(accuracy_score(y_test, clf.predict(X_test)))

# Creating a confusion matrix for the testing data
conf_mtrx_df = pd.DataFrame(confusion_matrix(y_test, clf.predict(X_test).reshape(-1, 1)))
conf_mtrx_df.rename(columns={0: 'Predicted (0)', 1: 'Predicted (1)'}, index={0: 'Target (0)', 1: 'Target (1)'})
```
